## Data manipulation

### The Basics of Pandas fillna() Function

##### The pandas fillna function is a powerful tool for handling missing data in pandas DataFrames. Its primary function is to fill in 'NaN' or missing values with a specified value or method

```python
# to replace NaN values with average rating 
# create a function using fillna 

def clean_nan(series):
    return series.fillna(series.median)
```

```python
# The transform() method allows you to execute a function for each value of the DataFrame.

store_df.Rating = store_df["Rating"].transform(clean_nan)
```

## Applying Lambda Functions to Pandas

 Syntax: lambda arguments: expression

An anonymous function which we can pass in instantly without defining a name or any thing like a full traditional function.

 Below are some methods and ways by which we can apply lambda functions to Pandas:

* Dataframe.assign() on a Single Column
* Dataframe.assign() on Multiple Columns
* Dataframe.apply() on a Single Row
* Dataframe.apply() on Multiple Rows
* Lambda Function on Multiple Rows and Columns Simultaneously

```python
store_df["Price"] = store_df["Price"].apply((lambda x:str(x).replace("$","")if "$"in str(x) else str(x)))
store_df["Price"] = store_df["Price"].apply(lambda x:float(x))
```

```python
store_df[475:485]
```

```python
store_df["Installs"] = store_df["Installs"].apply((lambda x:str(x).replace("+","")if "+"in str(x) else str(x)))
store_df["Installs"] = store_df["Installs"].apply((lambda x:str(x).replace(",","")if ","in str(x) else str(x)))

store_df["Installs"] = store_df["Installs"].apply(lambda x:float(x))
```
